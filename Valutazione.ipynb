```python
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
data = pd.read_csv('results.csv')
data
```

```python
data = data.groupby(['file','pel']).mean()
data
```

```python
data['total time'] = data['Phase 0']+data['Phase 1']+data['Phase 2']+data['Phase 3']+data['Phase 4']
data
```

```python
if not os.path.exists('Evaluation'):
    os.mkdir('Evaluation')


sns.set_theme(rc={'figure.figsize':(12,8)})
```

```python
sns.barplot(data=data, x='file', y='total time', hue='pel',errorbar=None)
plt.xlabel('Image')
plt.ylabel("ms")
plt.title("Evaluation " + 'total time')
plt.savefig("Evaluation/Evaluation total time.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
for col in data.columns[:5]:
    sns.barplot(data=data, x='file', y=col, hue='pel',errorbar=None)
    plt.xlabel('Image')
    plt.ylabel("ms")
    plt.title("Evaluation " + col)
    plt.savefig("Evaluation/Evaluation " + col+".png", dpi=300, bbox_inches='tight')
    plt.show()


for col in data.columns[5:10]:
    sns.barplot(data=data, x='file', y=col, hue='pel',errorbar=None)
    plt.xlabel('Image')
    plt.ylabel(col)
    plt.title("Evaluation " + col)
    if col == 'accuracy' or col == 'jaccard':
        plt.ylim(0,1)
    plt.savefig("Evaluation/Evaluation " + col+".png", dpi=300, bbox_inches='tight')
    plt.show()

```
